# 1. Retreiving data from the Aalto Open Courses API

```python
import json

# read key
with open('keys/keys.json') as f:
    keys = json.load(f)
    key = keys['aaltoapi']
```

```python
import requests

# call API
url = 'https://course.api.aalto.fi:443/api/sisu/v1/courseunitrealisations?user_key=' + key
r = requests.get(url)

# get data
data = r.json()
```

```python
# remove the courses that have languageOfInstruction = 'en'
print("There are", len(data), "courses in total.")
data = [course for course in data if 'en' in course['languageOfInstructionCodes']]
print("There are", len(data), "courses in English.")

# save data in a json file
with open('data/data.json', 'w') as f:
    json.dump(data, f)
```

```python
import pandas as pd

courses = pd.DataFrame(data)
courses.head()
```
